```python
# Exploratory Data Analysis on Titanic Dataset (train.csv)

# Step 1: Upload CSV Files
from google.colab import files
uploaded = files.upload()  # Upload train.csv, test.csv, gender_submission.csv

# Step 2: Import Required Libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

# Step 3: Load the Datasets
train_df = pd.read_csv("train.csv")
test_df = pd.read_csv("test.csv")
gender_df = pd.read_csv("gender_submission.csv")

# Step 4: Preview the Data
print("Train Data:")
print(train_df.head())

print("\nDataset Info:")
train_df.info()

print("\nSummary Statistics:")
print(train_df.describe(include='all'))

print("\nMissing Values:")
print(train_df.isnull().sum())

# Step 5: Value Counts for Key Categorical Columns
print("\nValue Counts:")
print(train_df['Sex'].value_counts())
print(train_df['Pclass'].value_counts())
print(train_df['Embarked'].value_counts())

# Step 6: Visualizations

# Age Distribution
plt.figure(figsize=(8, 5))
sns.histplot(train_df['Age'].dropna(), kde=True)
plt.title("Age Distribution")
plt.show()

# Gender Count
sns.countplot(x='Sex', data=train_df)
plt.title("Gender Distribution")
plt.show()

# Passenger Class Count
sns.countplot(x='Pclass', data=train_df)
plt.title("Passenger Class Distribution")
plt.show()

# Survival Count
sns.countplot(x='Survived', data=train_df)
plt.title("Survival Count (0 = No, 1 = Yes)")
plt.show()

# Survival by Gender
sns.countplot(x='Survived', hue='Sex', data=train_df)
plt.title("Survival by Gender")
plt.show()

# Survival by Class
sns.countplot(x='Survived', hue='Pclass', data=train_df)
plt.title("Survival by Passenger Class")
plt.show()

# Boxplot: Age by Gender
sns.boxplot(x='Sex', y='Age', data=train_df)
plt.title("Age Distribution by Gender")
plt.show()

# Boxplot: Age by Class
sns.boxplot(x='Pclass', y='Age', data=train_df)
plt.title("Age Distribution by Passenger Class")
plt.show()

# Step 7: Correlation Heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(train_df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()

# Step 8: Pairplot (selected features)
selected_cols = ['Age', 'Fare', 'Pclass', 'Survived']
sns.pairplot(train_df[selected_cols].dropna(), hue='Survived')
plt.suptitle("Pairplot of Age, Fare, Pclass and Survival", y=1.02)
plt.show()

# Step 9: Scatter Plot (Age vs Fare by Survival)
sns.scatterplot(x='Age', y='Fare', hue='Survived', data=train_df)
plt.title("Age vs Fare Colored by Survival")
plt.show()

# Step 10: Boxplot - Fare by Survival
sns.boxplot(x='Survived', y='Fare', data=train_df)
plt.title("Fare Distribution by Survival")
plt.show()

# Step 11: Summary of Findings
print("\nSummary of Key Findings:")
print("""
1. Most passengers were in 3rd class, and most did not survive.
2. Females had a much higher survival rate than males.
3. Younger passengers and 1st class passengers had better survival rates.
4. Fare and Pclass are positively correlated with survival.
5. The dataset has missing values in 'Age', 'Cabin', and 'Embarked'.
6. Fare distribution is right-skewed (some passengers paid very high fares).
""")
```
